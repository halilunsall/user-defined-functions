```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
def show_nulls(data):
    
    '''
    
    This function plots missing values for each column by observation in the dataset.
    
    ''' 
    plt.figure(figsize=(5, 5))

    sns.displot(data=data.isnull().melt(value_name="missing"),
                y="variable",
                hue="missing",
                multiple="fill",
                height=9.25)

    plt.axvline(0.2, color="r")
    plt.show();
```

```python
def fill_miss(data, col_num, agg_func, col_target, first_col, second_col=None, third_col=None):
    '''
    
    This function is created to fill in missing data according to similar properties of other columns.
    
    Parameters
    -----------
    
    dataframe : (pandas data frame)
    dataframe you want to fill nan values
    
    col_num : int
    the number of columns you want to group
    
    agg_func : str ('mean' or 'mode')
    aggregate function you want to implement
    
    col_target : str
    Column name where you want null values to be filled
    
    first_col : str
    the first column you want to set a grouping relationship to
    
    second_col : str
    second column you want to set grouping relation, if your col_num variable is 2 you need to fill here
    
    third_col : str
    third column you want to set grouping relation, if your col_num variable is 3 you need to fill here
    
    
    '''
    
    
    indexes = data[data[col_target].isnull()].index
    
#     return indexes
    
    for i in indexes:
        first_cond = data[first_col] == data[first_col][i]
        second_cond = third_cond = True
        
        if col_num >= 2:
            if second_col is not None:
                second_cond = data[second_col] == data[second_col][i]
            else:
                second_cond = True
        if col_num == 3:
            if third_col is not None:
                third_cond = data[third_col] == data[third_col][i]
            else:
                third_cond = True

        if agg_func == 'mean':
            mean_1 = data.loc[first_cond, col_target].mean()
            all_mean = data[col_target].mean()
            
            if col_num == 1:
                data.loc[i, col_target] = mean_1
                if np.isnan(data.loc[i, col_target]):
                    data.loc[i, col_target] = all_mean
            elif col_num == 2:
                mean_2 = data.loc[first_cond & second_cond, col_target].mean()
                data.loc[i, col_target] = mean_2
                if np.isnan(data.loc[i, col_target]):
                    data.loc[i, col_target] = mean_1
                    if np.isnan(data.loc[i, col_target]):
                        data.loc[i, col_target] = all_mean
            elif col_num == 3:
                mean_2 = data.loc[first_cond & second_cond, col_target].mean()
                mean_3 = data.loc[first_cond & second_cond & third_cond, col_target].mean()
                
                data.loc[i, col_target] = mean_3
                if np.isnan(data.loc[i, col_target]):
                    data.loc[i, col_target] = mean_2
                    if np.isnan(data.loc[i, col_target]):
                        data.loc[i, col_target] = mean_1
                        if np.isnan(data.loc[i, col_target]):
                            data.loc[i, col_target] = all_mean
                            
        elif agg_func == 'mode':
            mode_1 = data.loc[first_cond, col_target].mode()
            all_mode = data[col_target].mode()
            
            if col_num == 1: 
                if len(mode_1) > 0:
                    data.loc[i, col_target] = mode_1[0]
                else:
                    data.loc[i, col_target] = all_mode[0]
            elif col_num == 2:
                mode_2 = data.loc[first_cond & second_cond, col_target].mode()
                if len(mode_2) > 0:
                    data.loc[i, col_target] = mode_2[0]
                elif len(mode_1) > 0:
                    data.loc[i, col_target] = mode_1[0]
                else:
                    data.loc[i, col_target] = all_mode[0]
            elif col_num == 3:
                mode_2 = data.loc[first_cond & second_cond, col_target].mode()
                mode_3 = data.loc[first_cond & second_cond & third_cond, col_target].mode()
                if len(mode_3) > 0:
                    data.loc[i, col_target] = mode_3[0]
                elif len(mode_2) > 0:
                    data.loc[i, col_target] = mode_2[0]
                elif len(mode_1) > 0:
                    data.loc[i, col_target] = mode_1[0]
                else:
                    data.loc[i, col_target] = all_mode[0]
```
