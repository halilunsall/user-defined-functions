### Import Library's

```python
import pandas as pd
import regex as re
import warnings
pd.options.display.max_rows = 10

warnings.filterwarnings("ignore")
warnings.warn("this will not show")
```

### Function

```python
def organize_categoricals(data, target_col, prem_list, stan_list):
    '''
    This function is created to edit the excess unique values found in the data.
    '''
    
    
    for i in range(len(data)):
        premium, standart = 0, 0

        for prem in prem_list: 
            try:
                if prem == re.search(prem, data[target_col][i])[0]:
                    premium +=1
            except:
                pass
        for stan in stan_list: 
            try:
                if stan == re.search(stan, data[target_col][i])[0]:
                    standart +=1
            except:
                pass

        if premium == 0 and standart == 0:
            data[target_col][i] = data[target_col][i]
        else:
            try:
                if (100*premium)/(premium+standart) >= 50:
                    data[target_col][i] = 'Premium'
                else:
                    data[target_col][i] = 'Standart'
            except:
                data[target_col][i] = 'Standart'
```

##### The function groups premium and standard with simple mathematical operations.

Definition of the values we need to give to the function:
- data: your pandas dataframe you are working on
- target_col: column you want to edit
- prem_list = the first list created when you split the data in the column into two
    - Example: premium = ['Hilton Bathroom', 'Jacuzzi', 'Marble Floor', 'Furnished', 'Sauna', 'Terrace']
- stan_list = the second list created when you split the data in the column into two
    - Example: standart = ['ADSL', 'Wood Joinery', 'Built-in Kitchen', 'Balcony', 'Laundry Room']
    
            - Note: Make sure you get all the unique values in the column, otherwise you may not get good results.
